```python
#get initial data (stock prices)
price=pd.read_html('https://www.multpl.com/s-p-500-historical-prices/table/by-month')[0]
price['Date']=pd.to_datetime(price['Date'])
price['Date']=price['Date'].dt.strftime('%Y-%m')
price.set_index(price['Date'],inplace=True)
price
```

```python
price.info()
```

```python
# create dictionary with all data
m_data={}
for i in m_site:
  m_data[i[23:-15]+i[-9:]]=pd.read_html(i)[0]
  m_data[i[23:-15]+i[-9:]]['Date']=pd.to_datetime(m_data[i[23:-15]+i[-9:]]['Date'])
  m_data[i[23:-15]+i[-9:]]['Date']=m_data[i[23:-15]+i[-9:]]['Date'].dt.strftime('%Y-%m')
  m_data[i[23:-15]+i[-9:]].set_index(m_data[i[23:-15]+i[-9:]]['Date'],inplace=True)
  m_data[i[23:-15]+i[-9:]].drop(columns=['Date'],inplace=True)
```
